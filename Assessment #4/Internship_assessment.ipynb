```python
import numpy as np


def update_language_preferences(df):

    total_patients = len(df) #  = 15

    # Compute exact patient counts per language
    language_counts = [int(total_patients * p) for p in probabilities] # for e.g. p=0.45 int(15*0.45) = 6,3,2,1,1

    # Adjust rounding errors
    while sum(language_counts) < total_patients:
        max_index = np.argmax(language_counts)
        language_counts[max_index] += 1  # Add remaining patients to the largest group i.e. Tamil-8, Telugu-3, Malayalam-2, Hindi-1,English-1

    # Create shuffled language list
    language_list = [lang for lang, count in zip(languages, language_counts) for _ in range(count)]
    np.random.shuffle(language_list)

    # Assign to DataFrame
    df["language"] = language_list

# Update patient language preferences
update_language_preferences(patients_df)

# Display updated DataFrame
patients_df
```

```python
def assign_age_and_channel(df):
    ages = []
    channels = []

    total_tamil_speakers = (df["language"] == "Tamil").sum()
    tamil_elderly = int(total_tamil_speakers * 0.25)  # 25% of Tamil speakers are elderly

    tamil_elderly_assigned = 0  # Counter for assigned Tamil elderly

    for lang in df["language"]:
        if lang == "Tamil" and tamil_elderly_assigned < tamil_elderly:
            age = np.random.randint(60, 71)  # Tamil elderly (60-70 years)
            channel = "IVR"
            tamil_elderly_assigned += 1
        else:
            age = np.random.randint(15, 70)  # Other languages can be elderly randomly
            channel = "IVR" if age >= 60 else ("WhatsApp" if age <= 30 else "SMS")

        ages.append(age)
        channels.append(channel)

    df["age"] = ages
    df["channel"] = channels


assign_age_and_channel(patients_df)
patients_df
```

```python
from deep_translator import GoogleTranslator

def translate_message(message, target_language):
    """Translate message to the target language using Google Translate API"""
    translated = GoogleTranslator(source="auto", target=target_language).translate(message)
    return translated
```

```python
from gtts import gTTS
from IPython.display import Audio

# Language codes for gTTS
lang_codes = {
    "Tamil": "ta",
    "Telugu": "te",
    "Malayalam": "ml",
    "Hindi": "hi",
    "English": "en"
}

def generate_voice_message(text, language):
    """Generate and play voice message in Colab."""
    lang_code = lang_codes.get(language, "en")  # Default to English
    try:
        tts = gTTS(text=text, lang=lang_code)
        file_path = "message.mp3"
        tts.save(file_path)
        display(Audio(file_path))  # ‚úÖ Plays in Colab
    except ValueError as e:
        print(f"Error: {e} - Language not supported: {language}")
```

```python
def adapt_message_for_elderly(message):
    """Modify message format for elderly patients (larger font in apps)"""
    return f"{message.upper()}"  # Bold text, uppercase for emphasis
```

```python

# Function to Send Messages Based on Patient Preferences
def send_messages(message_type):
    """Send messages according to patient preferences"""
    for _, patient in patients_df.iterrows():
        language = patient["language"]
        channel = patient["channel"]
        message = message_templates.get(message_type, {}).get(language)

        # Auto-translate if template missing
        if message is None:
            message = translate_message("Your appointment is confirmed. Please visit!", language)

        # Elderly Patients (Modify Message)
        if patient["age"] >= 60:
            message = adapt_message_for_elderly(message)

        # IVR Patients (Convert to Voice)
        if channel == "IVR":
            bold_text = f"\033[1m{message}\033[0m"
            print(f"Calling {patient['name']} ({patient['channel']}) in {patient['language']}:\n {bold_text}\n")
            generate_voice_message(message, language)
        else:
            print(f"Sending via {channel} to {patient['name']} ({language}):\n {message}\n")


send_messages("AppointmentReminder")
```

```python
import random
import pandas as pd
import numpy as np

# Step 1: Define Patient Data (Sample Data)
patients_df = pd.DataFrame({
    "id": range(1, 16),
    "name": [f"Patient {i}" for i in range(1, 16)],
    "language": np.random.choice(["Tamil", "Telugu", "Malayalam", "Hindi", "English"], size=15),
    "channel": np.random.choice(["SMS", "WhatsApp", "IVR"], size=15),
})

# Step 2: Randomly Assign A/B Groups
patients_df["group"] = np.random.choice(["A", "B"], size=len(patients_df))

# Step 3: Send Messages Based on Group Assignment
confirmation_responses = []
survey_responses = []
for _, patient in patients_df.iterrows():
    if patient["group"] == "A":
        message = "Your appointment is confirmed. Please visit!"  # English Message
    else:
        message = f"[In {patient['language']}] Your appointment is confirmed. Please visit!"

    # Simulate random confirmation (1 for confirmed, 0 for no response)
    confirmation_responses.append(random.choice([0, 1]))

    # Simulate patient satisfaction survey response (1-5 rating)
    survey_responses.append(random.randint(1, 5))

    print(f"Sent to {patient['name']} ({patient['channel']} - {patient['group']}): {message}")

# Step 4: Measure Effectiveness
patients_df["confirmed"] = confirmation_responses
confirmation_rates = patients_df.groupby("group")["confirmed"].mean() * 100

print("\nConfirmation Rates:")
print(confirmation_rates)

# Step 5: Collect and Analyze Survey Responses
patients_df["survey_rating"] = survey_responses
survey_averages = patients_df.groupby("group")["survey_rating"].mean()

print("\nAverage Satisfaction Ratings:")
print(survey_averages)
```
